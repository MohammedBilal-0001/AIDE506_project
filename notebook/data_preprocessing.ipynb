```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
data_file=pd.read_csv('G:\LebanseUni\M2\S1\AIDE506-Advanced Machine learining Topics\Project\Data\WA_Fn-UseC_-Telco-Customer-Churn.csv')
data_file.head()
```

## check missing values

```python
print(data_file.dtypes)
```

```python
print((data_file["TotalCharges"] == " ").sum())
```

```python
data_file["TotalCharges"] = data_file["TotalCharges"].replace(" ", pd.NA)
```

```python
print(data_file.isnull().sum())
```

```python
data_file["TotalCharges"] = pd.to_numeric(data_file["TotalCharges"], errors='coerce')
data_file["TotalCharges"].fillna(data_file["TotalCharges"].median(), inplace=True)
```

## drop useless columns

```python
data_file.drop(columns=["customerID"], inplace=True)
```

## convert catigury to int

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}  # Dictionary to store LabelEncoder objects for each column

for column in data_file.columns:
    if data_file[column].dtype == "object":
        le = LabelEncoder()  # Create a new LabelEncoder instance for each column
        data_file[column] = le.fit_transform(data_file[column])
        label_encoders[column] = le  # Store the encoder for later use

        print(f"Column: {column}, Unique Values: {data_file[column].unique()}")

# Print stored mappings for each column
print("\nLabel Mappings:")
for column, le in label_encoders.items():
    print(f"{column}: {dict(zip(le.classes_, le.transform(le.classes_)))}")
```

```python
print("\nLabel Mappings for Encoded Columns:")
print(label_encoders)
for column, le in label_encoders.items():
    print(f"\nColumn: {column}")
    for original_label, encoded_value in zip(le.classes_, le.transform(le.classes_)):
        print(f"  {original_label} -> {encoded_value}")
```

## standardization

```python
scaler = StandardScaler()
data_file[["tenure", "MonthlyCharges", "TotalCharges"]] = scaler.fit_transform(data_file[["tenure", "MonthlyCharges", "TotalCharges"]])
```
